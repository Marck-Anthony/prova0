Realize o merge dos canais, mas troque a posição do R com o B para criar um efeito visual.

```python
import cv2
from google.colab.patches import cv2_imshow

# 1. Carregar a imagem original (Padrão BGR)
img = cv2.imread("/content/prova0/img/5.png")

# 2. Separar os canais individuais
(canalB, canalG, canalR) = cv2.split(img)

# 3. Realizar o merge trocando R com B
# Originalmente a ordem é [B, G, R].
# Vamos trocar para [R, G, B] para causar o efeito visual.
img_efeito = cv2.merge([canalR, canalG, canalB])

# 4. Exibir o resultado
print("Imagem com canais R e B trocados:")
cv2_imshow(img_efeito)

# 5. Opcional: Salvar o resultado
cv2.imwrite("/usr/img/imagem_efeito_troca.jpg", img_efeito)
```

Aumente o brilho da imagem somando 50 unidades aos pixéis (trate o transbordo com cv2.add).

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

# 1. Carregar a imagem original
img = cv2.imread("/content/prova0/img/8.png")

# 2. Criar uma matriz com o valor 50 para a soma
# Ela deve ter o mesmo tamanho e tipo da imagem original
M = np.ones(img.shape, dtype="uint8") * 50

# 3. Adicionar o brilho tratando o transbordo (overflow)
# A função cv2.add garante que o valor máximo seja 255
img_brilhante = cv2.add(img, M)

# 4. Exibir os resultados para comparação
print("Imagem Original:")
cv2_imshow(img)

print("Imagem com Brilho (+50):")
cv2_imshow(img_brilhante)
```

Redimensione duas fotos para o mesmo tamanho e faça a mesclagem com addWeighted (pesos 0.5/0.5)

```python
import cv2
from google.colab.patches import cv2_imshow

# 1. Carregar as duas imagens
img1 = cv2.imread("/content/prova0/img/7.png")
img2 = cv2.imread("/content/prova0/img/8.png")

# 2. Definir um tamanho padrão (ex: 500x500 pixels)
# No resize, a ordem é (largura, altura)
largura, altura = 500, 500
dimensoes = (largura, altura)

# 3. Redimensionar ambas para o mesmo tamanho
img1_redim = cv2.resize(img1, dimensoes, interpolation=cv2.INTER_AREA)
img2_redim = cv2.resize(img2, dimensoes, interpolation=cv2.INTER_AREA)

# 4. Mesclar as imagens com pesos iguais (0.5 cada)
# O último parâmetro (0) é o 'gamma', um valor escalar adicionado à soma
img_mesclada = cv2.addWeighted(img1_redim, 0.5, img2_redim, 0.5, 0)

# 5. Exibir o resultado
print("Resultado da mesclagem (Blending) 50/50:")
cv2_imshow(img_mesclada)
```

 Crie uma máscara binária circular e use-a para isolar uma parte da sua foto principal.

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

# 1. Carregar a imagem original
img = cv2.imread("/content/prova0/img/10.png")
(H, W) = img.shape[:2]

# 2. Criar uma máscara preta (zeros) com o mesmo tamanho da imagem
# Usamos apenas 1 canal para a máscara (tons de cinza)
mask = np.zeros((H, W), dtype="uint8")

# 3. Desenhar um círculo branco na máscara
# Parâmetros: (imagem, centro, raio, cor, espessura -1 para preencher)
centro = (W // 2, H // 2) # Centro da imagem
raio = 100
cv2.circle(mask, centro, raio, 255, -1)

# 4. Aplicar a máscara à imagem original usando BITWISE AND
# A função isola os pixels onde a máscara é branca
img_isolada = cv2.bitwise_and(img, img, mask=mask)

# 5. Exibir os resultados
print("Máscara Binária Circular:")
cv2_imshow(mask)

print("\nParte isolada da foto:")
cv2_imshow(img_isolada)
```

Calcule e imprima os valores de intensidade do histograma da sua imagem.

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# 1. Carregar a imagem original em tons de cinza
# O histograma de intensidade é mais simples de visualizar em um único canal
img = cv2.imread("/content/prova0/img/8.png", cv2.IMREAD_GRAYSCALE)

# 2. Calcular o histograma
# Parâmetros: [imagem], [canal], máscara, [tamanho_do_hist], [intervalo]
hist = cv2.calcHist([img], [0], None, [256], [0, 256])

# 3. Imprimir os valores numéricos de intensidade
print("Valores de intensidade (primeiros 10 níveis):")
for i in range(10):
    print(f"Intensidade {i}: {hist[i][0]} pixels")

# 4. Exibir o gráfico do histograma usando Matplotlib
plt.figure()
plt.title("Histograma de Intensidade (Tons de Cinza)")
plt.xlabel("Intensidade do Pixel")
plt.ylabel("Quantidade de Pixels")
plt.plot(hist)
plt.xlim([0, 256])
plt.show()
```

 Plote o histograma (gráfico) dos canais de cores separadamente no mesmo gráfico.

```python
import cv2
from matplotlib import pyplot as plt

# 1. Carregar a imagem colorida
img = cv2.imread("/content/prova0/img/7.png")

# 2. Definir as cores dos canais para o gráfico
# O OpenCV usa BGR, então definimos a tupla nesta ordem
cores = ('b', 'g', 'r')

plt.figure(figsize=(10, 6))
plt.title("Histograma por Canais de Cores (BGR)")
plt.xlabel("Intensidade do Pixel (0-255)")
plt.ylabel("Quantidade de Pixels")

# 3. Calcular e plotar o histograma para cada canal
for i, col in enumerate(cores):
    # Calcula o histograma para o canal 'i'
    # [i] representa o canal (0=Azul, 1=Verde, 2=Vermelho)
    hist = cv2.calcHist([img], [i], None, [256], [0, 256])

    # Plota no gráfico usando a cor correspondente
    plt.plot(hist, color=col, label=f'Canal {col.upper()}')
    plt.xlim([0, 256])

# 4. Exibir legenda e o gráfico final
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

Aplique a Equalização de Histograma para melhorar o contraste de uma imagem escura sua.

```python
import cv2
from google.colab.patches import cv2_imshow

# 1. Carregar a imagem original
img = cv2.imread("/content/prova0/img/6.png")

# 2. Converter para tons de cinza (essencial para equalização simples)
img_cinza = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# 3. Aplicar a Equalização de Histograma
img_equalizada = cv2.equalizeHist(img_cinza)

# 4. Exibir os resultados lado a lado para comparação
print("Original (Cinza) vs Equalizada:")
# Concatenar horizontalmente para facilitar a visualização
comparativo = cv2.hconcat([img_cinza, img_equalizada])
cv2_imshow(comparativo)
```

Aplique um limiar (Threshold) binário simples para separar o objeto do fundo
na sua foto.

```python
import cv2
from google.colab.patches import cv2_imshow

# 1. Carregar a imagem e converter para tons de cinza
# A binarização requer uma imagem de canal único
img = cv2.imread("/content/prova0/img/5.png")
img_cinza = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# 2. Aplicar o Threshold Binário Simples
# Parâmetros: imagem, valor do limiar, valor máximo, tipo de threshold
# O valor 128 é um ponto médio comum para começar os testes
(T, img_binaria) = cv2.threshold(img_cinza, 128, 255, cv2.THRESH_BINARY)

# 3. Exibir o resultado
print(f"Limiar utilizado: {T}")
cv2_imshow(img_binaria)
```
